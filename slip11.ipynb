```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = [
['butter','bread','milk'],
['butter','flour','milk','sugar'],
['butter','eggs','milk','salt'],
['eggs'],
['butter','flour','milk','salt']
]
print("Transactions Dataset:")
for i, t in enumerate(transactions,1):
    print("TID", i, ":", t)

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)
print("\nDataset in Numeric Format:")
print(df)
frequent_itemsets1 = apriori(df, min_support=0.4, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.4):")
print(frequent_itemsets1)

rules1 = association_rules(frequent_itemsets1, metric="confidence",
min_threshold=0.5)
print("\nAssociation Rules (min_support = 0.4):")
print(rules1)

frequent_itemsets2 = apriori(df, min_support=0.6, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.6):")
print(frequent_itemsets2)
rules2 = association_rules(frequent_itemsets2, metric="confidence",
min_threshold=0.5)
print("\nAssociation Rules (min_support = 0.6):")
print(rules2)
```
