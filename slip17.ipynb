```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from heapq import nlargest
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Step 1: Given Text Paragraph
text = """So, keep working. Keep striving. Never give up. Fall down seven times,
get up eight.
Ease is a greater threat to progress than hardship. Ease is a greater threat to
progress than hardship.
So, keep moving, keep growing, keep learning. See you at work."""
print("Original Text:\n", text)
# Step 2: Preprocess text (remove special characters and digits)
clean_text = re.sub('[^a-zA-Z ]', '', text)
print("\nCleaned Text:\n", clean_text)
# Step 3: Tokenize sentences
sentences = sent_tokenize(text)
# Step 4: Remove stopwords and create word frequency table
stop_words = set(stopwords.words('english'))
word_freq = {}
for word in word_tokenize(clean_text.lower()):
    if word not in stop_words:
        if word not in word_freq:
            word_freq[word] = 1
        else:
            word_freq[word] += 1
# Step 5: Score sentences
sent_scores = {}
for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in word_freq:
            if sent not in sent_scores:
                sent_scores[sent] = word_freq[word]
            else:
                sent_scores[sent] += word_freq[word]
# Step 6: Generate summary (Top 2 sentences)
summary_sentences = nlargest(2, sent_scores, key=sent_scores.get)
summary = ' '.join(summary_sentences)
print("\nGenerated Summary:\n", summary)
```
