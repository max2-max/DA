```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
['Bread','Milk'],
['Bread','Diaper','Beer','Eggs'],
['Milk','Diaper','Beer','Coke'],
['Bread','Milk','Diaper','Beer'],
['Bread','Milk','Diaper','Coke']
]
print("Transactions:")
for i, t in enumerate(dataset,1):
    print("TID",i,":",t)

te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
print("\nConverted Dataset (Numeric Format):")
print(df)

frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.4):")
print(frequent_itemsets)

rules = association_rules(frequent_itemsets, metric="confidence",
min_threshold=0.6)
print("\nAssociation Rules:")
print(rules)

frequent_itemsets2 = apriori(df, min_support=0.6, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.6):")
print(frequent_itemsets2)
rules2 = association_rules(frequent_itemsets2, metric="confidence",
min_threshold=0.6)
print("\nAssociation Rules (min_support = 0.6):")
print(rules2)
```
