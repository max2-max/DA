```python
Build a simple linear regression model for Fish Species Weight Prediction.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {
'Length': [23.2, 24.5, 25.3, 26.7, 27.5, 28.1, 29.0, 30.2, 31.5, 32.0],
'Weight': [242, 290, 340, 363, 430, 450, 500, 540, 600, 650]
}
df = pd.DataFrame(data)
print("Fish Dataset:")
print(df)

X = df[['Length']] 
y = df['Weight'] 
print("\nIndependent Variable (X):")
print(X)
print("\nTarget Variable (y):")
print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
random_state=0)
print("\nX_train:")
print(X_train)
print("\nX_test:")
print(X_test)
print("\ny_train:")
print(y_train)
print("\ny_test:")
print(y_test)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("\nPredicted Fish Weights:")
print(y_pred)
```
