```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('punkt')
nltk.download('vader_lexicon')
```

```python
df = pd.read_csv("covid_2021_1.csv")
print("Dataset Information:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())

df.dropna(inplace=True)

df.drop_duplicates(inplace=True) 
print("\nDataset after cleaning:")
print(df.info())

df['tokens'] = df['Comment'].astype(str).apply(word_tokenize)
print("\nTokenized Comments:")
print(df[['Comment','tokens']].head())

sia = SentimentIntensityAnalyzer()
df['sentiment_score'] = df['Comment'].astype(str).apply(lambda x:
sia.polarity_scores(x)['compound'])

df['sentiment'] = df['sentiment_score'].apply(
lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

total = len(df)
positive = len(df[df['sentiment'] == 'Positive'])
negative = len(df[df['sentiment'] == 'Negative'])
neutral = len(df[df['sentiment'] == 'Neutral'])
positive_percent = (positive/total)*100
negative_percent = (negative/total)*100
neutral_percent = (neutral/total)*100
print("\nTotal Comments:", total)
print("Positive Comments:", positive, "(", positive_percent, "% )")
print("Negative Comments:", negative, "(", negative_percent, "% )")
print("Neutral Comments:", neutral, "(", neutral_percent, "% )")
```
