```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("Groceries_dataset.csv", header=None)
print("Dataset Information:")
print(df.info())
print("\nFirst 5 Rows of Dataset:")
print(df.head())

df = df.dropna(how='all')
print("\nDataset after removing null values:")
print(df)

transactions = []
for i in range(len(df)):
    transactions.append([str(item) for item in df.iloc[i] if str(item) != 'nan'])
print("\nSample Transactions:")
print(transactions[:5])

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
data = pd.DataFrame(te_array, columns=te.columns_)
print("\nDataset in Numeric Format:")
print(data.head())

frequent_itemsets = apriori(data, min_support=0.02, use_colnames=True)
print("\nFrequent Itemsets:")
print(frequent_itemsets)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
print("\nAssociation Rules:")
print(rules)
```
