```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = {
    'company': ['Tata','MG','KIA','Hyundai'],
    'model': ['Nexon','Astor','Seltos','Creta'],
    'year': [2017,2021,2019,2015]  
}
df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

df = df.drop(columns=['year'])

df_numeric = pd.get_dummies(df)
print("\nDataset after converting categorical values to numeric:")
print(df_numeric)

frequent_itemsets1 = apriori(df_numeric, min_support=0.5, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.5):")
print(frequent_itemsets1)

if not frequent_itemsets1.empty:
    rules1 = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.5)
    print("\nAssociation Rules (min_support = 0.5):")
    print(rules1)

frequent_itemsets2 = apriori(df_numeric, min_support=0.25, use_colnames=True)
print("\nFrequent Itemsets (min_support = 0.25):")
print(frequent_itemsets2)

if not frequent_itemsets2.empty:
    rules2 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.5)
    print("\nAssociation Rules (min_support = 0.25):")
    print(rules2)
```
